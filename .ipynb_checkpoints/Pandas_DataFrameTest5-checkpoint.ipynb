```python
import pandas as pd
import numpy as np
```

```python
data = [{'eng':30,'kor':70,'math':100},
       {'eng':40,'kor':80,'math':10},
       {'eng':50,'kor':90,'math':20},
       {'eng':60,'kor':100,'math':30},
       {'eng':70,'kor':10,'math':40}]
df =pd.DataFrame(data,index=['aa','bb','cc','dd','ee'])
df
```

```python
# 특정 범위를 지정하여 출력
df[(df['eng']>=30)&(df['kor']<=50)]
```

# 추가, 수정, 삭제, 검색, 정렬

```python
# 해당 컬럼명이 있으면 수정, 없으면 추가됨
df['my']=[1,2,3,4,5]
df
```

```python
# eng컬렴의 값으로 새로운 컬럼이 생성
df['my1']=df['eng']
df
```

```python
# 각 컬럼의 요소값을 더해서 새로운 컬럼이 생성
df['my2']=df['eng']+df['math']
df
```

```python
df.loc['aa']=[1,2,3]
df
```

```python
# 새로운 행 추가
df.loc['ff']=[1,2,3]
df
```

```python
# 삭제 
# drop(index=[]) 해당 인덱스(행) 삭제
df.drop(index=['aa','dd'],inplace=True)
df
```

```python
# drop(columns=[]) 해당 컬럼(열) 삭제
df.drop(columns=['eng'])
df
```

## NaN 데이터

```python
# 데이터가 비어있으면 NaN(np.nan)으로 표시
df['my']=[10,20,np.nan,np.nan,30]
df
```

```python
# nan(결측치) 삭제
# 특정 (평균값)값으로 채워준다.
```

```python
# nan인것은 True로 표시
# df.isna()

# 컬럼별 nan(결측치)의 갯수를 계산
df.isna().sum()
```

```python
# nan 데이터 삭제 nan 데이터가 있는 행 삭제
df.dropna()
```

```python
# nan 데이터 삭제 nan 데이터가 있는 행 삭제
df.dropna(axis=1) # 결측치가 있는 컬럼(열)을 삭제
```

```python
df.dropna(axis=0) # 결측치가 있는 인덱스(행)를 삭제
```

```python
# nan의 값을 ()안의 값으로 바꿔줌
# df.fillna(10)

# 특정 컬럼의 nan값을 ()안의 값으로 바꿔줌
df['my']=df['my'].fillna(20)
df
```

## 정렬

```python
# (by=' ')안에 기준 컬럼을 적어줘야함
df.sort_values(by='kor')
```
