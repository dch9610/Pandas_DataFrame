```python
import pandas as pd
import numpy as np
```

```python
data = [{'eng':30,'kor':70,'math':100},
       {'eng':40,'kor':80,'math':10},
       {'eng':50,'kor':90,'math':20},
       {'eng':60,'kor':100,'math':30},
       {'eng':70,'kor':10,'math':40}]
df =pd.DataFrame(data,index=['aa','bb','cc','dd','ee'])
df
```

```python
df+1
```

```python
df*2
```

```python
# df[컬럼명] = [...] 해당 컬럼값 수정
df['eng']=[1,2,3,4,5]
df
```

```python
# df.loc['컬럼명'] = df.loc['컬럼명'] + a  해당 컬럼에만 연산을 적용
df['eng']=df['eng']+2
df
```

```python
# df.loc['행이름'] = df.loc['행이름'] +a 특정 행에만 연산적용
df.loc['aa'] = df.loc['aa']+2
df
```

```python
# row boolean indexing, True에 해당하는 로우만 출력
df[[True,False,True,False,True]]
```

```python
# 해당 조건을 만족하는 열의 값을 출력하고 
df[df['eng']>=50]
```
