```python
import pandas as pd
import numpy as np
```

```python
# https://pandas.pydata.org/pandas-docs/stable/reference/groupby.html
# Groupby Document 사이트
```

```python
df=pd.read_excel('data/hotel.xlsx',index_col='Unnamed: 0')
df
```

```python
# ~별 groupby
# DataFrameGroupBy 멤버(집계)  sum, mean, max, min, agg
g=df.groupby('grade')
g.sum()[['price']] #집계대상컬럼  정수, 실수로된 모든 컬럼이 대상
```

```python
g=df.groupby('grade') # grade로 그룹을 만듬
g.mean().round(2)[['price']] # 소수점 2자리까지만 출력
```

```python
g=df.groupby('grade') # grade로 그룹을 만듬
g.max().round(2)[['price']] # 소수점 2자리까지만 출력
```

```python
g=df.groupby('grade') # grade로 그룹을 만듬
g.count()[['price']] # 소수점 2자리까지만 출력
```

```python
g=df.groupby('grade') # grade로 그룹을 만듬
g.agg(['sum','mean']) # 한 칼럼에 sum과 mean 값이 나오게 함
```

```python
g=df.groupby('grade') # grade로 그룹을 만듬
g.agg(['sum','mean']).round(2)['price']
```
