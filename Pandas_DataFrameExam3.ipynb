```python
import pandas as pd
import numpy as np
import seaborn as sb

import matplotlib.pyplot as plt # 시각화 지원
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False
plt.rcParams["figure.figsize"]=(12,5)
```

```python
# sep=',' 인자를 구분된 /, .등으로 분류해서 읽어옴
# 첫번째 로우는 컬럼명으로 지정하지 않겠다. header =None
# index_col은 앞부분 컬럼을 인덱스로 빼줌
# 인코딩 에러 'euc-kr'

crimeDF =pd.read_csv('data/crime_in_Seoul.csv',encoding='euc-kr',thousands=',')
crimeDF
```

# 1번 살인발생이 가장 높은 관서명, 살인발생, 살인검거를 출력하시오.

```python
num = crimeDF['살인 발생'].max()
crimeDF[crimeDF['살인 발생']==num][['관서명','살인 발생','살인 검거']]
```

# 2번 관서별 절도발생, 절도 검거에 대한 바차트 그리시오.

```python
crimeDF1 =pd.read_csv('data/crime_in_Seoul.csv',encoding='euc-kr',thousands=',', index_col='관서명')


mydf = crimeDF1[['절도 발생','절도 검거']]
mydf.plot(kind='bar')
plt.show()
```

# 3번 절도검거율 컬럼을 추가하시오.

```python
crimeDF['절도 검거율(%)']=crimeDF['절도 검거']/crimeDF['절도 발생']*100
crimeDF
```

#  4번 
# 살인발생, 강도발생 2개 컬럼 평균값이 15건 이상이면 '범죄 매우 높음', 10건 이상이면 '범죄 높음' 미만이면 '보통' 표시되는 '강력 범죄 정도' 컬럼을 추가하시오.

```python
crimeDF['godam'] = (crimeDF['살인 발생']+crimeDF['강도 발생'])/2
crimeDF['강력범죄정도'] = crimeDF['godam'].apply(lambda x:'범죄매우높음' if x>=15 else '범죄높음' if x>=10 else '보통')
crimeDF
```

# 5. 관서명 마지막 글자 '서'를 제외한 글자로 구성된 '지역' 칼럼을 추가하시오.

```python
crimeDF['지역']=crimeDF['관서명'].str[0:-1]
crimeDF
```

# 6. 강도발생 탑5 관서명, 강도발생, 강도검거를 출력하시오.

```python
top5=crimeDF['강도 발생'].nlargest(5).min()
top=crimeDF[crimeDF['강도 발생']>=top5]
top[['관서명','강도 발생','강도 검거']].sort_values(by='강도 발생', ascending=False)
```

# 7. 살인발생, 강도발생, 절도발생의 상관관계를 구하고 heatmap으로 표시하시오.

```python
sbCR=crimeDF[['살인 발생','강도 발생','절도 발생']].corr()
sbCR
```

```python
sb.heatmap(sbCR,vmin=-1,vmax=1,annot=True,linewidths=0.5)
plt.show()
```
